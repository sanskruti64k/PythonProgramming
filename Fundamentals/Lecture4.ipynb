```python
# Looping and Idendation
i = 0
while i < 4:
  print(i)
  if i == 1:
    print ("Hello")
  i  = i + 1
```

```python
# Looping and Idendation
# Infinite loop without stop statements
i = 0
while i <= 5:
  if i == 1:
    print ("Hello")
    if i == 2:
      print("World")
      i  = i + 2
  print(i)
```

```python
#Calculator
username = input("Enter name :")
password = input("Enter password :")

if username.lower() == "harsh" and password == "1234":
  num1 = float(input("Enter number 1: "))
  num2 = float(input("Enter the value of number 2: "))
  operator = input("Enter the operation to be performed ")
  if operator == "+":
    print(num1 + num2)
  elif operator == "-":
    print(num1 - num2)
  elif operator == "*":
    print(num1 * num2)
  elif operator == "/":
    print(num1 / num2)
  elif operator == "%":
    print(num1 % num2)
  elif operator == "**":
    print(num1 ** num2)
  else:
    print("Invalid operation ")
```
